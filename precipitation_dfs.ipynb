```python
import pandas as pd
import dask.dataframe as dd
from datetime import datetime
```

## Creating dfs by century

### Make df for 2000+ only

```python
df = dd.read_csv('drought1_9/*.csv', usecols=['STATION','DATE','PRCP'])  # read all csv files in directory lazily
df.compute().to_csv('out2000.csv', index=False)  # convert to pandas and save as csv
```

```python
df_2000 = pd.read_csv('out2000.csv')
```

```python
df_2000['DATE'] = pd.to_datetime(df_2000['DATE'])
```

### Make df for 1900-1999

```python
df = dd.read_csv('drought11_48/*.csv', usecols=['STATION','DATE','PRCP'])  # read all csv files in directory lazily
df.compute().to_csv('out1900.csv', index=False)  # convert to pandas and save as csv
```

```python
df_1900 = pd.read_csv('out1900.csv')
```

```python
df_1900['DATE'] = pd.to_datetime(df_1900['DATE'])
```

```python
df_1900['DATE'] = df_1900['DATE'].dt.strftime('19%y-%m-%d')
```

### Make df for 1800-1899

```python
df = dd.read_csv('drought50_52/*.csv', usecols=['STATION','DATE','PRCP'])  # read all csv files in directory lazily
df.compute().to_csv('out1800.csv', index=False)  # convert to pandas and save as csv
```

```python
df_1800 = pd.read_csv('out1800.csv')
```

```python
df_1800['DATE'] = pd.to_datetime(df_1800['DATE'])
```
